We now extract pairs of (text,label) from our TED data

```python
with zipfile.ZipFile('ted_en-20160408.zip', 'r') as z:
    doc = lxml.etree.parse(z.open('ted_en-20160408.xml', 'r'))
input_keywords = '\n'.join(doc.xpath('//head/keywords/text()'))
input_text = doc.xpath('//content/text()')
input_text[5][0:100]
input_text[296]
```

Next we format the input keywords correctly, ensuruing that each talk is represented by one symbol of the set {ooo,Too,oEo,ooD,TEo,oED,ToD,TED}. 

```python
labels_ted = []
for labels in input_keywords.split('\n'):
    labels_ted.append(labels.split(', '))
tech_labels = []
edu_labels = []
des_labels = []
for talk_labels in labels_ted:
    if 'technology' in talk_labels:
        tech_labels.append('T')
    else:
        tech_labels.append('o')
    if 'education' in talk_labels:
        edu_labels.append('E')
    else:
        edu_labels.append('o')
    if 'design' in talk_labels:
        des_labels.append('D')
    else:
        des_labels.append('o')
labels = [tech_labels[i]+edu_labels[i]+ des_labels[i] for i in range(len(tech_labels))]
print(labels[1:15])
```

```python
documents = list(zip(input_text,labels))
(documents[5][0][0:50]+'...',documents[5][1])
```

<h1> The model </h1>

First we split our data into training and test 

```python
from random import shuffle
print(len(documents))
shuffle(documents)
training = documents[0:1585]
validation = documents[1585:1585+250]
test = documents[1585+250:1585+500]
```

Next we tokenise all the talks into words within sentences

```python
def tokenise(text):
    sentences = []
    #remove all paraenthis
    re.sub(r'\([^)]*\)', '', text)
    for line in text.split('\n'):
        m = re.match(r'^(?:(?P<precolon>[^:]{,20}):)?(?P<postcolon>.*)$', line)
        sentences.extend(sent for sent in m.groupdict()['postcolon'].split('.') if sent)
    talk = []
    for sentence in sentences:
        tokens = re.sub(r"[^a-z0-9]+", " ", sentence.lower()).split()
        talk.append(tokens)
    return talk

def sentences(text):
    sentences_tokens = []
    for talk in text:
        sentences_tokens.append((tokenise(talk[0]),talk[1]))
    return sentences_tokens
training_sentences_tokens = sentences(training)    
training_sentences_tokens[5][0][0][0:10]
```

Now we use wordToVec on our training data to get the word encodings we desire

```python
#This is the eventual size of the vectors representing documents as well
embedding_size = 50
embedding_window = 5
embedding_min_count = 50

from gensim.models import Word2Vec
#here we need to append all the training talk's sentences
model_input = [['UNK' for n in range(embedding_min_count)]]
for talk in training_sentences_tokens:
    model_input.extend(talk[0])
model = Word2Vec(model_input, size=embedding_size, window = embedding_window, min_count = embedding_min_count, workers=4)
```

<b> We now lose our sentence stucture as we make each talk a vector of word embeddings </b>

```python
def to_talk_tokens(data):
    return [[[word for sentence in talk[0] for word in sentence],talk[1]] for talk in data]

training_talks = to_talk_tokens(training_sentences_tokens)
training_talks[5][0][0:10]
```

```python
def generate_word_matrix(word_embedding):
    word_dict = dict()
    words = list()
    for i, word in enumerate(model.vocab):
        word_dict[word] = i
        words.append(model[word])
    return word_dict, np.asarray(words)
w2v_dict, w2v_matrix = generate_word_matrix(model)
```

```python
def embed_words(talk,word_token_dict):
    talk_into_tokens = np.zeros(5403)
    for i in range(min(len(talk),5403)):
        talk_into_tokens[i] = word_token_dict.get(talk[i],word_token_dict['UNK'])
    return talk_into_tokens
```

```python
import random 
def training_data_generator(batch_size, word_token_dict):
    talks_into_tokens = [(embed_words(talk[0],word_token_dict),label_dict[talk[1]]) for talk in training_talks]
    while True:
        random.shuffle(talks_into_tokens)
        for i in range(len(talks_into_tokens)//batch_size):
            batch = talks_into_tokens[50*i:50*i+50]
            yield [np.array([np.array(shit[0]) for shit in batch]),np.array([shit[1] for shit in batch])]
```
